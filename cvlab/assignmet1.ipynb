```python
from PIL import Image
import numpy as np

# Open an image file
with Image.open("C:\\Users\\SUJAY.J\\Desktop\\cvprojects\\1.jpg") as img:
    # Convert the image to grayscale (optional, but simplifies calculations)
    img = img.convert('L')
    
    # Convert the image to a NumPy array
    img_array = np.array(img)

    # Calculate the area (number of pixels)
    area = img_array.size

    # Calculate mean, min, and max pixel values
    mean_val = np.mean(img_array)
    min_val = np.min(img_array)
    max_val = np.max(img_array)

# Print the results
print(f"Area (number of pixels): {area}")
print(f"Mean pixel value: {mean_val}")
print(f"Minimum pixel value: {min_val}")
print(f"Maximum pixel value: {max_val}")
```

```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt

# Load an image
image_path = "C:\\Users\\SUJAY.J\\Desktop\\cvprojects\\1.jpg"  # Replace with your image path
image = Image.open(image_path)

# Transformations
flipped_image = image.transpose(Image.FLIP_LEFT_RIGHT)
inverted_image = ImageOps.invert(image)
rotated_90_image = image.rotate(90)
rotated_left_image = image.rotate(90, expand=True)
rotated_right_image = image.rotate(-90, expand=True)

# Display the images in a Jupyter notebook
fig, axs = plt.subplots(1, 5, figsize=(25, 5))

# Original image
axs[0].imshow(image)
axs[0].set_title('Original Image')
axs[0].axis('off')

# Flipped image
axs[1].imshow(flipped_image)
axs[1].set_title('Flipped Image')
axs[1].axis('off')

# Inverted image
axs[2].imshow(inverted_image)
axs[2].set_title('Inverted Image')
axs[2].axis('off')

# Rotated left image
axs[3].imshow(rotated_left_image)
axs[3].set_title('Rotated Left')
axs[3].axis('off')

# Rotated right image
axs[4].imshow(rotated_right_image)
axs[4].set_title('Rotated Right')
axs[4].axis('off')

plt.show()
```

```python
from PIL import Image, ImageFilter
import numpy as np
import matplotlib.pyplot as plt

# Open an image file
with Image.open("C:\\Users\\SUJAY.J\\Desktop\\cvprojects\\3.jpg") as img:
    # Apply a Gaussian blur filter
    blurred_img = img.filter(ImageFilter.GaussianBlur(radius=5))  # Adjust the radius for more or less blur
    
    # Apply a contour filter
    contour_img = img.filter(ImageFilter.CONTOUR)  # Apply contour filter
    
    # Apply a detail filter
    detail_img = img.filter(ImageFilter.DETAIL)  # Apply detail filter
    
    # Function to calculate and plot histogram
    def plot_histogram(image, ax, title):
        # Convert image to grayscale for histogram
        gray_image = image.convert("L")  # Convert to grayscale
        histogram, bins = np.histogram(np.array(gray_image).flatten(), bins=256, range=[0, 256])
        ax.plot(histogram, color='black')  # Plot histogram
        ax.set_title(title)
        ax.set_xlim([0, 256])  # Set x-axis limits
        ax.set_ylim([0, max(histogram) + 10])  # Set y-axis limits

    # Display the original and filtered images in the notebook
    plt.figure(figsize=(15, 10))  # Set the figure size for images and histograms

    # Image Subplots
    plt.subplot(2, 4, 1)  # 2 rows, 4 columns, 1st subplot
    plt.title("Original Image")
    plt.imshow(img)  # Display original image
    plt.axis('off')  # Hide axes

    plt.subplot(2, 4, 2)  # 2 rows, 4 columns, 2nd subplot
    plt.title("Blurred Image")
    plt.imshow(blurred_img)  # Display blurred image
    plt.axis('off')  # Hide axes

    plt.subplot(2, 4, 3)  # 2 rows, 4 columns, 3rd subplot
    plt.title("Contour Image")
    plt.imshow(contour_img)  # Display contour image
    plt.axis('off')  # Hide axes

    plt.subplot(2, 4, 4)  # 2 rows, 4 columns, 4th subplot
    plt.title("Detail Image")
    plt.imshow(detail_img)  # Display detail image
    plt.axis('off')  # Hide axes

    # Histogram Subplots
    plt.subplot(2, 4, 5)  # 2 rows, 4 columns, 5th subplot
    plot_histogram(img, plt.gca(), "Histogram - Original")

    plt.subplot(2, 4, 6)  # 2 rows, 4 columns, 6th subplot
    plot_histogram(blurred_img, plt.gca(), "Histogram - Blurred")

    plt.subplot(2, 4, 7)  # 2 rows, 4 columns, 7th subplot
    plot_histogram(contour_img, plt.gca(), "Histogram - Contour")

    plt.subplot(2, 4, 8)  # 2 rows, 4 columns, 8th subplot
    plot_histogram(detail_img, plt.gca(), "Histogram - Detail")

    plt.tight_layout()  # Adjust layout to prevent overlap
    plt.show()  # Show the plots
```

```python
import matplotlib.pyplot as plt  # Fixed the import statement
import cv2  # Assuming you're using OpenCV to read images

def plot_histogram(img, ax, title):
    # Calculate histogram
    histogram = cv2.calcHist([img], [0], None, [256], [0, 256])
    ax.plot(histogram, color='black')  # Plot histogram
    ax.set_title(title)
    ax.set_xlim([0, 256])  # Set x-axis limits
    ax.set_ylim([0, max(histogram) + 10])  # Set y-axis limits

# List of image file paths
image_paths = ["C:\\Users\\SUJAY.J\\Desktop\\cvprojects\\1.jpg", "C:\\Users\\SUJAY.J\\Desktop\\cvprojects\\2.jpg", "C:\\Users\\SUJAY.J\\Desktop\\cvprojects\\3.jpg",
               "C:\\Users\\SUJAY.J\\Desktop\\cvprojects\\4.jpg", "C:\\Users\\SUJAY.J\\Desktop\\cvprojects\\5.jpg", "C:\\Users\\SUJAY.J\\Desktop\\cvprojects\\6.jpg",
               "C:\\Users\\SUJAY.J\\Desktop\\cvprojects\\7.jpg", "C:\\Users\\SUJAY.J\\Desktop\\cvprojects\\8.jpg", "C:\\Users\\SUJAY.J\\Desktop\\cvprojects\\9.jpg",
               "C:\\Users\\SUJAY.J\\Desktop\\cvprojects\\10.jpg"]

# Create a figure to display images and histograms
plt.figure(figsize=(20, 15))  # Increase the figure size for a wider layout

for idx, img_path in enumerate(image_paths):
    img = cv2.imread(img_path)  # Read the image
    # Image Subplots
    plt.subplot(4, 10, idx + 1)  # 4 rows, 10 columns for images
    plt.title(f"Image {idx + 1}")
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Display original image
    plt.axis('off')  # Hide axes

    # Histogram Subplots
    plt.subplot(4, 10, idx + 11)  # 4 rows, 10 columns for histograms (adjusted index)
    plot_histogram(img, plt.gca(), f"Histogram - Image {idx + 1}")

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()  # Show the plots
```
